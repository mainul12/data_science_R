````r
library(tidyverse)
````

````r
1:10 %>%
    map(rnorm, n = 2)
````

````r
data <- 1:10 %>%
  map(function(x) rnorm(100, x, x))
````

```R
1:3 %>%
    map(function(x) hist(data[[x]]))
```

````r
1:10 %>%
    map(~ rnorm(3, .x, .x))
````

````r
data %>%
    map_dbl(mean)
````
